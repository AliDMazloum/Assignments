```python
import numpy as np
import matplotlib.pyplot as plt
import math

plt.style.use(["science", "notebook", "grid"])


ax = plt.axes()

# Data for a three-dimensional line
xline = np.linspace(-3, 3, 1000)
yline = [ (3 - math.sqrt(i**2))/3 for i in xline]
plt.plot(xline,yline)

yline = [ (math.sqrt(36 -  4*i**2))/9 for i in xline]
yline2 = [ -(math.sqrt(36 -  4*i**2))/9 for i in xline]

plt.xlabel("x1")
plt.ylabel("x2")

plt.title('Part B: 1')
plt.plot(xline,yline,color="orange")
plt.plot(xline,yline2,color="orange")
# plt.savefig("PartB_1.pdf")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
xline = np.linspace(0, 6, 1000)
yline = xline 
zline = xline 
ax.plot3D(xline, yline, zline, 'orange')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')


plt.title('Part B: 2.1')
# plt.savefig("PartB_2.1.pdf")
```

```python
# create the figure
fig = plt.figure()

xline = np.linspace(-3, 3, 100)
yline = np.linspace(-3, 3, 100)



# add axes
ax = fig.add_subplot(111,projection='3d')

xx, yy = np.meshgrid(xline, yline)
zline = 1 - xx - yy

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

# plot the plane
ax.plot_surface(xx, yy, zline, alpha=0.5)

plt.title('Part B: 2.2')
# plt.savefig("PartB_2.2.pdf")

# plt.show()
```

```python
# create the figure
fig = plt.figure()

xline = np.linspace(-3, 3, 100)
yline = np.linspace(-3, 3, 100)



# add axes
ax = fig.add_subplot(111,projection='3d')

xx, yy = np.meshgrid(xline, yline)
zline = xx**2 + yy**2

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

# plot the plane
ax.plot_surface(xx, yy, zline, alpha=0.5)

plt.title('Part B: 3.1 (it is a convex)')
plt.savefig("PartB_3.1.pdf")

# plt.show()
```

```python
# create the figure
fig = plt.figure()

xline = np.linspace(0.1, 3, 100)
yline = np.linspace(0.1, 3, 100)

xlinelog = [math.log(i) for i in xline]
ylinelog = [math.log(i) for i in yline]



# add axes
ax = fig.add_subplot(111,projection='3d')

xx, yy = np.meshgrid(xline, yline)
xxlog, yylog = np.meshgrid(xlinelog, ylinelog)
zline = -xx*(xxlog)  - yy*yylog

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

# plot the plane
ax.plot_surface(xx, yy, zline, alpha=0.5)

plt.title('Part B: 3.2 (it is not a convex)')
plt.savefig("PartB_3.2.pdf")

# plt.show()
```

```python
# create the figure
fig = plt.figure()

xline = np.linspace(-3, 3, 100)
yline = np.linspace(-3, 3, 100)

def abs(x):
    for i in range(len(x)):
        for j in range(len(x[i])):
            if x[i][j] < 0:
                x[i][j] = - x[i][j]
    return x



# add axes
ax = fig.add_subplot(111,projection='3d')

xx, yy = np.meshgrid(xline, yline)

zline = abs(xx) +  abs(yy)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

# plot the plane
xline = np.linspace(-3, 3, 100)
yline = np.linspace(-3, 3, 100)
xx, yy = np.meshgrid(xline, yline)

ax.plot_surface(xx, yy, zline, alpha=0.5)

plt.title('Part B: 3.3 (it is a convex)')
plt.savefig("PartB_3.3.pdf")

# plt.show()
```
