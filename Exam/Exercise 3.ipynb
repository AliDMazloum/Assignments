```python
import numpy as np
import math
import time
import matplotlib.pyplot as plt
```

```python
def f(coeff, x):
    return (coeff[0] + coeff[1]*(math.sin(x) + math.cos(x)) + coeff[2]*(math.sin(2*x) + math.cos(2*x)) + coeff[3]*(math.sin(3*x) + math.cos(3*x)))
def cost_function(coeff,x,y):
    cost = 0
    for i in range(len(x)):
        cost += (f(coeff, x[i])-y[i])**2
    return cost
def direction_d(coeff,x,y):
    d0 = 0
    for i in range(len(x)):
        d0 += 2*(f(coeff, x[i])-y[i])
    
    d1 = 0
    for i in range(len(x)):
        d1 += 2*(math.sin(x[i])+math.cos(x[i]))*(f(coeff, x[i])-y[i])
        
    d2 = 0
    for i in range(len(x)):
        d2 += 2*(math.sin(2*x[i])+math.cos(2*x[i]))*(f(coeff, x[i])-y[i])
        
    d3 = 0
    for i in range(len(x)):
        d3 += 2*(math.sin(3*x[i])+math.cos(3*x[i]))*(f(coeff, x[i])-y[i])
    
    return np.array([-d0,-d1,-d2,-d3])

def print_row(counter, x, d, a, f="s"):
    if f =="s":
        print("k\t","x1\t", "x2\t", "d1\t", "d2\t", "||d||\t", "a\t", "f(x)\t")
    else:        
        x1 = round(x[0][0],6)
        x2 = round(x[1][0],6)
        d1 = round(d[0][0],6)
        d2 = round(d[1][0],6)
        d_norm = round(np.linalg.norm(d),6)
        a = round(a[0],6)
        f = round(f[0],6)
        print(counter,"\t", x1,"\t", x2,"\t", d1,"\t", d2,"\t", d_norm,"\t", a,"\t", f)
        
X = np.arange(-1,1.1,0.1)
Y = [-0.96,-0.577, -0.073, 0.377, 0.641, 0.66, 0.461, 0.134, -0.201, -0.434, -0.5, -0.393, -0.165, 0.099, 0.307,0.396, 0.345, 0.182, -0.031, -0.219, -0.321]
x_current = np.array([[1],[1],[1],[1]])
x_values = [x_current]
d_current = direction_d(x_current, X, Y)
a_current = np.array([0.017])
apsilon = 10**(-3)
counter = 0

print_row(counter, x_current, d_current, a_current, f="s")

while (np.linalg.norm(d_current) > apsilon):
    print_row(counter, x_current, d_current, a_current, cost_function(x_current, X, Y))
    x_new = x_current + a_current*d_current
    d_new = direction_d(x_new, X, Y)
    
    x_current = x_new
    x_values.append(x_current)
    d_current = d_new
    counter+=1
```

```python
plt.plot(X,Y, label="Given graph")
fitted = [f(x_current, i) for i in X]
plt.plot(X,fitted, label="Fitted graph")
plt.title("Exercise 3")
plt.legend()
plt.grid(True)
plt.savefig("Exercise3.pdf")
```

```python
plt.plot([i for i in range (len(x_values))],[cost_function(i, X, Y) for i in x_values])
plt.title("Iteration vs Cost")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.grid(True)
plt.savefig("Exercise3_costvsiteration.pdf")
```
