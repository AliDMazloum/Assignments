```python
import numpy as np
import math
import time
import matplotlib.pyplot as plt
```

```python
t = [-2,-1,0,1,2]
y = [2,-10,0,2,1]

def p(t):
    return ((-1 - 14/8) +t+(14/16)*t**2 )
```

```python
fitted = [p(i) for i in t]
print (fitted)
```

```python
plt.plot(t,y,label = "Given graph")
plt.plot(t,fitted, label = "Fitted Graph")
plt.title("Exercise 2")
plt.legend()
plt.savefig("Exercise2.pdf")
```
