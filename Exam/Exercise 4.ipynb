```python
import numpy as np
import matplotlib.pyplot as plt
import math

def circle_eq(x1, x2):
    return (x1 - 3)**2 + (x2 - 3)**2

ax = plt.axes()

# Data for a three-dimensional line
xline = np.linspace(-1, 3, 1000)
yline_upper = [ (math.sqrt(4 - (4/9)*i**2)) for i in xline]
yline_lower = [ -(math.sqrt(4 - (4/9)*i**2)) for i in xline]
# plt.plot(xline,yline)

xline_1 = np.linspace(-math.sqrt(1), math.sqrt(3), 1000)
yline2_upper = [ (math.sqrt(1 -  (1/3)*i**2)) for i in xline_1]
yline2_lower = [ -(math.sqrt(1 -  (1/3)*i**2)) for i in xline_1]

plt.xlabel("x1")
plt.ylabel("x2")

plt.title('Exercise 4')
plt.plot(xline,yline_upper,color="orange")
plt.plot(xline,yline_lower,color="orange")
plt.plot(xline_1,yline2_upper,color="orange")
plt.plot(xline_1,yline2_lower,color="orange")


x_c, y_c = 3, 3  
r_values = np.linspace(2, 2.9, 5) 
for r in r_values: 
    t = np.linspace(0, 2*np.pi, 100)
    x = x_c + r * np.cos(t)
    y = y_c + r * np.sin(t)

    # Plot the circle
    plt.plot(x, y,color="green")
    plt.scatter(1.3, 0.6,color="red")
    plt.axis('equal')
    plt.grid(True)

# plt.savefig("PartB_1.pdf")
```
